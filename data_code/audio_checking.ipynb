```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
from IPython.display import Audio, display
from plot_audios import *
```

```python
def plot_random_mel_spectrograms_one_by_one(folder_path, n=9):
    """
    Plot spectrograms for n random audio files from a folder one by one.
    
    Parameters:
    folder_path - Path to the folder containing audio files
    n           - Number of random audio files to plot (default is 9)
    """
    # Get list of all audio files in the folder
    audio_files = [f for f in os.listdir(folder_path) if f.endswith(('.wav', '.mp3', '.flac', '.ogg'))]
    
    if len(audio_files) < n:
        raise ValueError(f"The folder contains less than {n} audio files.")
    
    # Select n random audio files
    random_files = random.sample(audio_files, n)
    
    # Plot spectrograms one by one
    for file_name in random_files:
        file_path = os.path.join(folder_path, file_name)
        y, sr = librosa.load(file_path,sr=None)
        plot_mel_spectrogram(y, sr, file_name)
        display(Audio(y, rate=sr, autoplay=True))
```
